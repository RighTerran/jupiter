# Задача 3

#### Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока. Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями

```python
sns.boxplot(x=df['price'], y=df['waterfront'].astype(str), showfliers=False);
```

#### вид на набережную влияет на стоимость. В зависимости от есть ли вид или нет стоимость растет

```python
sns.boxplot(x=df['price'], y=df['view'].astype(str), showfliers=False);
```

#### насколько хороший вид из квартиры влияет на стоимость. Чем лучше вид, тем выше стоимость

```python
sns.boxplot(x=df['price'], y=df['grade'].astype(str), showfliers=False);
```

#### качество конструкции и дизайна влияет на стоимость, чем оно выше, тем выше стоимость

```python
corr_matrix= df.corr()
corr_matrix=np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix)<0.3]=0
corr_matrix
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, linewidths=1, cmap='coolwarm');
```

#### Наибольшое влияение на стоимость оказывает жилая площадь (0.7), конструкция и дизайн (0.7), кв. метры дома, которые находятся выше земли и Кв. метры жилой площади у 15 соседей
